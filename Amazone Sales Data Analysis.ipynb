## Checking corelation of ecah Column


```python
# Convert necessary columns to numeric if they're not already
columns_to_convert = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
       'Order Date', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
       'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit', 'Year',
       'Month', 'YearMonth']
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Calculate the correlation matrix for the relevant columns
correlation_matrix = data[['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
       'Order Date', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
       'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit', 'Year',
       'Month', 'YearMonth']].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix for Amazon Sales Data')
plt.show()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%Y')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%m/%d/%Y')

# Extract year and month from 'Order Date' for trend analysis
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['YearMonth'] = data['Order Date'].dt.to_period('M')
```

## Convert 'Order Date' and 'Ship Date' to datetime

```python
# Aggregate total revenue by month
monthly_sales = data.groupby(data['Order Date'].dt.month)['Total Revenue'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Month-wise Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=0)
plt.show()
```

##  Aggregate total revenue by month

```python
# Aggregate total sales by year
yearly_sales = data.groupby('Year')['Total Revenue'].sum()

plt.figure(figsize=(10, 5))
yearly_sales.plot(kind='bar', color='green')
plt.title('Year-wise Total Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

## Aggregate total sales by year
